```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

data['Final_Result'] = data['Final_Result'].map({'Fail': 0, 'Pass': 1})
data['Internet'] = data['Internet'].map({'No': 0, 'Yes': 1})

data = data.dropna()

output_dataset = data['Final_Result']
input_dataset = data[['GPA', 'Entrance_Test_Score', 'Internet']]

result_model = LogisticRegression()
result_model.fit(input_dataset, output_dataset)

insert = [[3.5, 55, 0]]
columns = ['GPA', 'Entrance_Test_Score', 'Internet']
column_df = pd.DataFrame(insert, columns=columns)

result_predictions = result_model.predict(column_df)

for i in result_predictions:
    if(i==0):
      print('Fail')
    else:
      print('Pass')
```
